```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
```

```python
users = pd.read_csv('Users.csv')
ratings = pd.read_csv('Ratings.csv')
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book_Rating':'num_rating'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['Book-Rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

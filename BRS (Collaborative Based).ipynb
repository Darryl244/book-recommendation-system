```python
import numpy as np
import pandas as pd
books = pd.read_csv('Books.csv')
users = pd.read_csv('Users.csv')
ratings = pd.read_csv('Ratings.csv')
ratings_with_name=ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
reg_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(reg_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
popular_books=y[y].index
popular_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(popular_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi_score=cosine_similarity(pt)
```

```python
simi_score.shape
```

```python
def recommend(book_name):

    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(simi_score[index])),key=lambda x:x[1],reverse=True)[1:6]
   
    for i in similar_items:
        print(pt.index[i[0]])
        
```

```python
recommend('Message in a Bottle')
```
